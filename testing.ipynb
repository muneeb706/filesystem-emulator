```python
from filesystem import FileSystem
from session import Session

fs = FileSystem()
fs.format()

fs.add_user('john')
fs.add_user('david')

session = Session(fs)
session.login('david')

while True:
  user_input = input(f'{session.user_name}@usc:{session.pwd()}$ ')

  if (user_input == 'exit'):
    break

  args = user_input.split()
  n_args = len(args)

  if (n_args == 0):
    continue

  cmd = args[0]

  if cmd == 'ls':
    res = session.ls()
    if res is not None:
      print(res)
  elif cmd == 'dump_fsimage':
    res = session.dump_fsimage()
    if res is not None:
      print(res)
  elif cmd == 'pwd':
    if n_args != 1:
      print('invalid input!')
    else:
      print(session.pwd())
  elif cmd == 'mkdir':
    if n_args != 2:
      print('invalid input!')
    else:
      session.mkdir(args[1])
  elif cmd == 'rmdir':
    if n_args != 2:
      print('invalid input!')
    else:
      session.rmdir(args[1])
  elif cmd == 'cd':
    if n_args == 1: # just cd
      session.cd()
    elif n_args == 2:
      session.cd(args[1])
    else:
      print('invalid input!')
  elif cmd == 'touch':
    if n_args == 2:
      session.touch(args[1])
    else:
      print('invalid input!')
  elif cmd == 'rm':
    if n_args == 2:
      session.rm(args[1])
    else:
      print('invalid input!')
  else:
    print('unknown command!')

session.dump_fsimage()
```
